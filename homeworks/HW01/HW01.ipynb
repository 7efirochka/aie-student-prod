```python
import numpy as np
```

### Создание массивов

```python
n1 = np.array([1, 2, 3])     # создание одномерного массива
print(n1)
```

```python
n2 = np.array([[1, 1, 1] , [2, 2, 2], [3, 3, 3]])    # создание двумерного массива
print(n2)
```

```python
n3 = np.array([[[1, 2, 1] , [2, 3, 2], [3, 4, 3]], [[4, 5, 4], [5, 6, 5], [6, 7, 6]]])    # создание трехмерного массива
print(n3)
```

```python
zer = np.zeros((2, 6, 5), int)    # матрица из нулей 
print(zer)
```

```python
one = np.ones((3, 5))    # матрица из единиц
print(one)
```

```python
arang = np.arange(2, 23, 4).reshape((2, 3))   # создание одномерной матрицы от 2 до 23 с шагом 4 и изменение её в двумерную с размером (2, 3)
print(arang)
```

```python
emp = np.empty((3, 6))
print(emp)
```

### Получение размера матрицы

```python
print(n2.shape)
print(zer.shape)
print(arang.shape)
```

### Получение типа данных матрицы

### Индексирование матрицы

```python
n3 = np.array([[[1, 2, 1] , [2, 3, 2], [3, 4, 3]], [[4, 5, 4], [5, 6, 5], [6, 7, 6]], [[7, 8, 7], [8, 9, 8], [9, 0 , 9]]])    # создание трехмерного массива
print(n3[1,2,1])  #  получние элемента
```

```python
n3 = np.array([[[1, 2, 1] , [2, 3, 2], [3, 4, 3]], [[4, 5, 4], [5, 6, 5], [6, 7, 6]], [[7, 8, 7], [8, 9, 8], [9, 0 , 9]]])    # создание трехмерного массива
print(n3)
print("-------------------------------")
print(n3[1:, :2, 1:2])  # получение подмассива
```

### Векторные операции с массивами

```python
a = np.array([1, 3, 5, 7])
b = np.array([2, 4, 6, 8])
print(a+b)
print(a*b)
print(a**3)
```

```python
print(a.mean())
print(a.min())
print(a.max())
```

```python
b = b.reshape((2, 2))
print(b)
print(b.min())  # минимальное значение матрицы
print(b.min(axis = 1)) # минимальное значение по строкам
```
